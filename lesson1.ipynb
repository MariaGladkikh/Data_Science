```python
%%time
#Easy
#Нужно написать функцию, которая на вход получает список из чисел, 
#а на выходе дает список, где каждое число возведено в квадрат. 
#Решать при помощи цикла.
list_e = list(range(100))
print (list_e)
m = len(list_e)
for i in range (m):
    print(i**2, end=" ")
```

```python
%%time
#Normal
#Тоже самое, но на этот раз без цикла (смотрим раздел с функциями из ноутбука, который разбирали в классе).

nsq_list = list(range(100))
print(nsq_list)
sq_list = [value ** 2 for value in range(100)]
print(sq_list)
```
